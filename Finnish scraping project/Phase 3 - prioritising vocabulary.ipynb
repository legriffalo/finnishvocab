```python
import pandas as pd
```

```python
freq = pd.read_excel('Frequency List.xlsx','The List - Frequency List')
freq = freq[1:]
freq.columns =['Summary', 'word', 'POS', 'freq', 'forms']
freq.head()

eng_verbs = freq[freq['POS']=='v']
eng_verbs.head()
```

```python
ranked = eng_verbs.sort_values('freq', ascending = False)
ranked['important'] = 'bonus'

ranked['important'][0:100]='critical'
ranked['important'][100:250]='very useful'
ranked['important'][250:500]='useful'
ranked['important'][500:750]='ok'
```
