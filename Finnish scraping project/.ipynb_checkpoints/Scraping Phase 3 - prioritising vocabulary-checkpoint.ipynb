```python
freq = pd.read_excel('Frequency List.xlsx','The List - Frequency List')
freq = freq[1:]
freq.columns =['Summary', 'word', 'POS', 'freq', 'forms']
freq.head()
```

```python
ranked = freq.sort_values('freq', ascending = False).copy()
ranked


ranked['important'] = 'bonus'

ranked['important'][0:250]='vital'
ranked['important'][250:500]='critical'
ranked['important'][500:2000]='extremely useful'
ranked['important'][2000:3000]='useful'
ranked['important'][3000:4000]='common speech'
ranked['important'][4000:]='bonus'
```
