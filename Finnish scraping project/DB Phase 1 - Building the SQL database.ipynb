```python
import pandas as pd
import os
```

```python
nouns = pd.read_csv('./cleaneddb/finnish_nouns.csv')
adjectives = pd.read_csv('./cleaneddb/finnish_adjectives.csv')
infinitives = pd.read_csv('./cleaneddb/verbinfinitives.csv')
```

```python
# now to union the verb tables from scraping phase 1 
files = os.listdir('./cleaneddb/verbs')
verbs = pd.DataFrame()

for file in files:
    f = pd.read_csv(f'./cleaneddb/verbs/{file}')
    
    verbs = pd.concat([verbs,f], ignore_index=True)
    
```

```python
verbs.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
verbs.shape
```

```python
verbs.head(10)
```

```python
verbs['person_eng'] = ''
verbs['person_eng'][verbs['person'].str.contains('minä')==True] = 'I'
verbs['person_eng'][verbs['person'].str.contains('sinä')==True] = 'You'
verbs['person_eng'][verbs['person'].str.contains('hän')==True] = 'he/she'
verbs['person_eng'][verbs['person'].str.contains('me')==True] = 'We'
verbs['person_eng'][verbs['person'].str.contains('te')==True] = 'you (pl/formal)'

verbs['person_eng'][verbs['person'].str.contains('he')==True] = 'they'
verbs['person_eng'][verbs['person'].str.contains('pass')==True] = 'passive'
```
